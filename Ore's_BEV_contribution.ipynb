```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

mydata =  pd.read_csv("BEV.csv")

mydata.head()
```

__Treating Null Values__

```python
mydata.drop(columns = ['CO2 emissions (g/km)','CO2 rating'], inplace = True)

mydata.head()
```

```python
mydata[mydata.isna().any(axis=1)]
```

```python
mydata['Smog rating'].isna().sum()
```

__Imputing null values with the average value__

```python
mydata['Smog rating'] = mydata['Smog rating'].fillna(mydata['Smog rating'].mean()) 

mydata['Smog rating'].isna().sum()
```

```python
mydata.isna().sum()
```

__NO NULL VALUES LEFT__

__Next, we went on to check for any spelling inconsistencies and check if there is any need for standardization in the categorical variables__

```python
mydata['Make'].unique()
```

```python
mydata.columns = mydata.columns.str.strip()
```

```python
mydata['Vehicle class'].unique()
```

```python
mydata['Model year'].dtype
```

__Which vehicle manufacturers have shown the highest improvement in range and efficiency over time?__

```python
## First i will start with computing efficiency 
```

```python
mydata['Efficiency'] = round((mydata['Range (km)'] / mydata['Combined (kWh/100 km)']),1)
```

```python
mydata.head()
```

```python
## Next is average range 
```

```python
avg_range_by_year = (
    mydata.groupby('Model year')['Range (km)']
    .mean()
    .reset_index()
    .sort_values(by='Model year')
)


plt.figure(figsize=(12,6))
plt.plot(avg_range_by_year['Model year'], avg_range_by_year['Range (km)'], marker='o', linestyle='-', color='green')

plt.title('Average BEV Range(km) Over Time (2012â€“2025)', fontsize=13)
plt.xlabel('Model Year')
plt.ylabel('Average Range (km)')
plt.grid(True, linestyle='--', alpha=0.5)


plt.text(avg_range_by_year['Model year'].iloc[0], avg_range_by_year['Range (km)'].iloc[0] + 10,
         f"{avg_range_by_year['Range (km)'].iloc[0]:.1f} km", color='black')
plt.text(avg_range_by_year['Model year'].iloc[-1], avg_range_by_year['Range (km)'].iloc[-1] + 10,
         f"{avg_range_by_year['Range (km)'].iloc[-1]:.1f} km", color='black')

plt.show()
```

__The average range of BEVs increased from 108.5km (2012) to 473.1km (2024), a 336.1% improvement â€” reflecting major gains in battery density and energy optimization.__

### Recharge Time 

```python
avg_recharge_by_year = (
    mydata.groupby('Model year')['Recharge time (h)'].mean().reset_index().sort_values(by='Model year')
)

plt.figure(figsize=(10,6))
plt.plot(avg_recharge_by_year['Model year'], avg_recharge_by_year['Recharge time (h)'],
         marker='o', linestyle='-', color='orange')

plt.title('Average Recharge Time of BEVs (2012â€“2025)', fontsize=13)
plt.xlabel('Model Year')
plt.ylabel('Recharge Time (hours)')
plt.grid(True, linestyle='--', alpha=0.5)


plt.text(avg_recharge_by_year['Model year'].iloc[0],
         avg_recharge_by_year['Recharge time (h)'].iloc[0] + 0.2,
         f"{avg_recharge_by_year['Recharge time (h)'].iloc[0]:.1f}h", color='black')
plt.text(avg_recharge_by_year['Model year'].iloc[-1],
         avg_recharge_by_year['Recharge time (h)'].iloc[-1] + 0.2,
         f"{avg_recharge_by_year['Recharge time (h)'].iloc[-1]:.1f}h", color='black')

plt.show()

```

```python
periods = []


for year in mydata['Model year']:
    if 2012 <= year <= 2017:
        periods.append('2012â€“2017')
    elif 2018 <= year <= 2025:
        periods.append('2018â€“2025')
    else:
        periods.append(np.nan)


mydata['Periods'] = periods

mydata.head()
```

```python
periods_comparison = (mydata.groupby(['Make', 'Periods'])['Combined (kWh/100 km)'].mean().reset_index())

Percentage_change = periods_comparison.pivot(index='Make',columns='Periods',values='Combined (kWh/100 km)')

Percentage_change.columns.name = None
Percentage_change = Percentage_change.reset_index()

Percentage_change.head()
```

__As seen above, there are a number of null values and this is because not all car Makes have models for every year. For the efficiency of this analysis, we would be removing the makes without both time periods__

```python
Percentage_change = Percentage_change.dropna()
```

```python
 Percentage_change['% change'] = round(((Percentage_change['2018â€“2025'] - Percentage_change['2012â€“2017']) / Percentage_change['2012â€“2017'])*100,2)
```

```python
Percentage_change = Percentage_change.sort_values(by = '% change')

Percentage_change
```

__Evidently Tesla made the most improvement towards energy efficiency__

```python
colors = ['green' if x < 0 else 'red' for x in Percentage_change['% change']]

# Plot
plt.figure(figsize=(10, 6))
plt.barh(Percentage_change['Make'], Percentage_change['% change'], color=colors)
plt.axvline(0, color='black', linewidth=1)  # vertical line at 0
plt.title('Change in Energy Consumption (kWh/100km) by Manufacturer', fontsize=13)
plt.xlabel('Percentage Change (%)')
plt.ylabel('Manufacturer')

plt.tight_layout()
plt.show()
```

__WHAT CAR MAKES THE MOST OF 1 UNIT OF ENERGY__

```python
efficiency_rate = mydata.groupby(by = ['Make'])['Efficiency', ].mean().reset_index()
```

```python
efficiency_rate = round(efficiency_rate.sort_values(by = 'Efficiency', ascending=False),1)

highest_efficiency_rate = efficiency_rate[:5]
highest_efficiency_rate
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

# Plot the pie chart
ax.pie(highest_efficiency_rate['Efficiency'], labels= highest_efficiency_rate['Make'],explode=(0.2, 0, 0, 0,0), shadow=True)
ax.set_title('TOP 5 CARS WITH THE HIGHEST EFFICIENCY RATE')
ax.legend(title="Make", loc="lower right")
plt.show()
```
